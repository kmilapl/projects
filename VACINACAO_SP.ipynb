# Importando Bibliotecas e CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('datasets/vacin.csv', encoding='utf-8')
```

# Volume da População por Município

```python
plt.figure(figsize=(8, 4))
plt.plot(dataset['municipio'], dataset['população'], marker="o",color='mediumpurple')
plt.title('Maiores Municípios de São Paulo de acordo com a População', color='blueviolet')
plt.legend(["População"], 
           bbox_to_anchor = (1.0, 0.13))
plt.gca().invert_yaxis() #Invertendo valores para ordem crescente
plt.grid(True)

#Criando tabela pra legendar os municipios que estão em siglas
fig, ax = plt.subplots()
table_data=[
    ["SP", 'São Paulo'],
    ["GRU", 'Guarulhos'],
    ["CAMP", 'Campinas'],
    ["SBC", 'São Bernardo do Campo'],
    ["SJC", 'São José dos Campos'],
    ["SA", 'Santo André'],
    ["RP", 'Ribeirão Preto'],
    ["OSAS", 'Osasco'],
    ["SOROC", 'Sorocaba'],
    ["MAUÁ", 'Mauá']
]
table = ax.table(cellText=table_data, loc='left')
table.set_fontsize(9)
table.scale(0.8,1)
ax.axis('off') #Para não apresentar um gráfico ao lado da tabela
plt.show()
```

# Número de Vacinação por Municipios

```python
plt.figure(figsize=(10, 5))
plt.plot(dataset['municipio'], dataset['dose_unica'], marker="s",color='#d0f4de', label='dose única')
plt.plot(dataset['municipio'], dataset['dose_um'], marker="o",color='#ff99c8', label='primeira dose')
plt.plot(dataset['municipio'], dataset['dose_dois'], marker="x",color='#a9def9', label='segunda dose')
plt.plot(dataset['municipio'], dataset['dose_adicional'], marker="D",color='#e4c1f9', label='dose adicional')
plt.title('DOSES - VACINAÇÃO COVID-19', color='black')
plt.legend(bbox_to_anchor = (1.0, 0.3))
plt.gca().invert_yaxis() #Invertendo valores para ordem crescente
plt.grid(True)

plt.figure(figsize=(10, 5))
plt.bar(dataset['municipio'], dataset['dose_unica'], color='#d0f4de', label='dose única')
plt.bar(dataset['municipio'], dataset['dose_um'], color='#ff99c8', label='primeira dose')
plt.bar(dataset['municipio'], dataset['dose_dois'], color='#a9def9', label='segunda dose')
plt.bar(dataset['municipio'], dataset['dose_adicional'], color='#e4c1f9', label='dose adicional')
plt.legend(bbox_to_anchor = (1.0, 1.0))
plt.show()
```

<h3> Com esses gráficos, podemos reparar que a dose única (que substitui a primeira e segunda dose) foi a que teve menor quantidade.<br>
Em SP, podemos reparar que a primeira dose foi a mais utilizada, assim como nos outros estados.
A diferença que podemos observar da primeira e segunda dose, é em MAUÁ, onde as duas ficam praticamente empatadas.</h3>
